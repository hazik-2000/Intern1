# droping the non required coulmns axis=1

```python
df=df.drop(["Summary","Precip Type","Wind Speed (km/h)","Wind Bearing (degrees)","Visibility (km)",'Pressure (millibars)',"Daily Summary"],axis=1)
```

```python
df
```

# changing type of formatted date from str to Timestamp

```python
df["Formatted Date"]=pd.to_datetime(df['Formatted Date'],utc=True)
```

```python
type(df['Formatted Date'])
```

```python
type(df['Formatted Date'][0])
```

# setting the index

```python
df.set_index("Formatted Date",inplace=True)
```

```python
df.head()
```

# Resampling data as per requirement

```python
df1=df[['Apparent Temperature (C)',"Humidity"]].resample('MS').mean()
```

```python
df1
```

```python
df1.head()
```

# PLOTTING FIRST PLOT FOR ANALYSIS OF TEMP VS HUMIDITY OVER A PERIOD OF YEAR

```python
plt.figure(figsize=(16,9))
plt.plot(df1['Humidity'],label='humidity')
plt.plot(df1['Apparent Temperature (C)'],label='Apparent temperature')
plt.xlabel("Years",fontsize=20)

plt.legend(shadow=True,ncol=2,labelspacing=3)
plt.title('Apparent Temperature VS Humidity 10years data',fontsize=20,bbox={'color':'silver'},pad=15)
plt.grid()
plt.show()
```
